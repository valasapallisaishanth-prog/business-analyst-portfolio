```python
import zipfile
import os
zip_file = list(uploaded.keys())[0]
with zipfile.ZipFile(zip_file, 'r') as zip_ref:
    zip_ref.extractall("loan_data")
print("Extracted files:", os.listdir("loan_data"))
```

```python
import pandas as pd
file_path = "loan_data/train_u6lujuX_CVtuZ9i.csv"  # update if needed
df = pd.read_csv(file_path)
df.head()
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_").str.lower()
df.head()
```

```python
print("Total Applications:", len(df))
print("Approval Rate:", df['loan_status'].value_counts(normalize=True)['Y'] * 100, "%")
print("Rejected Rate:", df['loan_status'].value_counts(normalize=True)['N'] * 100, "%")
```

```python
import plotly.express as px
px.pie(df, names="loan_status", title="Loan Approval Distribution")
```

```python
px.histogram(df, x="applicantincome", color="loan_status",
             title="Income Distribution by Loan Approval")
```

```python
fig3 = px.scatter(
    df,
    x="applicantincome",
    y="loanamount",
    color="loan_status",
    title="Income vs Loan Amount by Approval Status",
    hover_data=["education", "dependents"]
)
fig3.show()
```

```python
fig4 = px.histogram(
    df,
    x="credit_history",
    color="loan_status",
    barmode="group",
    title="Impact of Credit History on Loan Approval"
)
fig4.show()
```

```python
total = len(df)
approved = df['loan_status'].value_counts().get('Y', 0)
rejected = df['loan_status'].value_counts().get('N', 0)
print("Total:", total, "| Approved:", approved, "| Rejected:", rejected)
```

```python
summary = pd.DataFrame({
    "Metric": ["Total Applications", "Approved", "Rejected", "Approval Rate %"],
    "Value": [
        len(df),
        approved,
        rejected,
        round((approved/total)*100, 2)
    ]
})
summary
```

```python
fig5 = px.bar(
    summary,
    x="Metric",
    y="Value",
    title="Loan Application Summary Metrics",
    text="Value"
)
fig5.show()
```

```python
from graphviz import Digraph

# AS-IS Diagram
asis = Digraph('AS_IS_Loan_Process', format='png')
asis.attr(rankdir='LR')

asis.node('A', 'Customer Submits Paper\nApplication')
asis.node('B', 'Documents Received\n(Physical or Email)')
asis.node('C', 'Manual Document\nVerification')
asis.node('D', 'Are all\ndocuments valid?', shape='diamond')
asis.node('E', 'Credit Officer Review')
asis.node('F', 'Request Resubmission')
asis.node('G', 'Manual Decision\n& Notify Customer')

asis.edge('A', 'B')
asis.edge('B', 'C')
asis.edge('C', 'D')
asis.edge('D', 'E', label='Yes')
asis.edge('D', 'F', label='No')
asis.edge('E', 'G')

asis.render(filename="AS_IS_Loan_Process", directory="/content", cleanup=True)


# TO-BE Diagram
tobe = Digraph('TO_BE_Loan_Process', format='png')
tobe.attr(rankdir='LR')

tobe.node('A', 'Customer Portal\n(Online Form)')
tobe.node('B', 'Auto Document\nValidation')
tobe.node('C', 'Credit Scoring Engine')
tobe.node('D', 'Rule-Based Decision\nEngine', shape='diamond')
tobe.node('E', 'Auto Approval')
tobe.node('F', 'Escalate to\nCredit Officer')
tobe.node('G', 'Dashboard Tracking\n& KPIs')
tobe.node('H', 'Automated SMS/Email\nNotification')

tobe.edge('A', 'B')
tobe.edge('B', 'C')
tobe.edge('C', 'D')
tobe.edge('D', 'E', label='Approve')
tobe.edge('D', 'F', label='Escalate')
tobe.edge('E', 'G')
tobe.edge('F', 'G')
tobe.edge('G', 'H')

tobe.render(filename="TO_BE_Loan_Process", directory="/content", cleanup=True)

print(" Diagrams generated successfully!")
```
